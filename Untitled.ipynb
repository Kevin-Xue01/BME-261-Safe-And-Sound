```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture(0)
while True:
    _, frame = cap.read()
    cv2.imshow('Frame', frame)
    
    key = cv2.waitKey(1)
    if key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
img = cv2.imread('assets/main_view.jpg', 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(img)
```

```python
blur = cv2.GaussianBlur(img,(5,5),0)
plt.axis("off")
plt.imshow(blur)
```

```python
blur_bi = cv2.bilateralFilter(img,37,75,75)

bw = cv2.cvtColor(blur_bi, cv2.COLOR_RGB2GRAY)
thresh_ad = cv2.adaptiveThreshold(bw,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,61,2)

plt.axis("off")
plt.imshow(thresh_ad)
plt.show()
```

```python
# tried opening (erosion then dilation) to rmv noise but didn't work
# made inv of threshold so that features are white

#kernel_open = np.ones(bw.shape,np.uint8)
thresh_inv = cv2.bitwise_not(thresh_ad)
#opening = cv2.morphologyEx(thresh_inv, cv2.MORPH_OPEN, kernel_open)

kernel_erode = np.ones((3,3),np.uint8)
erosion = cv2.erode(thresh_inv,kernel_erode,iterations = 6)

kernel_dilate = np.ones((3,3),np.uint8)
dilation = cv2.dilate(erosion,kernel_dilate,iterations = 6)

plt.axis("off")
plt.imshow(thresh_inv)
plt.show()

plt.axis("off")
plt.imshow(erosion)
plt.show()

plt.axis("off")
plt.imshow(dilation)
plt.show()
```
